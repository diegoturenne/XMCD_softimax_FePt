```python
class STXM_data:
    '''
    Common class to store a SXTM dataset
    
    Complaints and curses to be directed to: Diego Turenne
    '''
    
    
    ## FIND COORDINATES AND ENERGIES
    def get_stack_coordinates(self, metadata_file):
        '''
        takes a metadata filename (string) 
        returns:
        x_coordinates, y_coordinates, energy_coordinates
        '''
    

        file_metadata = open(metadata_file,'r')
        metadata = file_metadata.read()


        # X positions
        match = re.compile('PAxis = {(.|\n)*?[}][;][\n]').search(metadata) # matches everything after PAxis
        tmp_x = match.group(0)

        match = re.compile('[(](.)*[)]').search(tmp_x)
        tmp_x = match.group(0)
        tmp_x = tmp_x[1:-1]
        x_pos = np.asarray(tmp_x.split(',')).astype(float)
        nb_xpts = x_pos[0]
        x_pos = x_pos[1::]

        # Y positions
        match = re.compile('QAxis = {(.|\n)*?[}][;][\n]').search(metadata) # matches everything after PAxis
        tmp_y = match.group(0)

        match = re.compile('[(](.)*[)]').search(tmp_y)
        tmp_y = match.group(0)
        tmp_y = tmp_y[1:-1]
        y_pos = np.asarray(tmp_y.split(',')).astype(float)
        nb_ypts = y_pos[0]
        y_pos = y_pos[1::]

        # Energies
        match = re.compile('StackAxis = {(.|\n)*?[}][;][\n]').search(metadata) # matches everything after PAxis
        tmp_e = match.group(0)

        match = re.compile('[(](.)*[)]').search(tmp_e)
        tmp_e = match.group(0)
        tmp_e = tmp_e[1:-1]
        energies = np.asarray(tmp_e.split(',')).astype(float)
        nb_energypts = energies[0]
        energies = energies[1::]
        
        return x_pos, y_pos, energies

        
    def load_data(self,data_folder, meta_file, x_pos, y_pos, energies ):
        '''
        loads the actual images into the file system using the metadata, metadata needs to be loaded.
        creates self.image_stack with all them images 
        
        retruns stack of images with shape((len(energies), len(x_pos), len(y_pos))
        '''
        image_stack = np.zeros((len(energies), len(x_pos), len(y_pos)) )

        for i, energy in enumerate(self.energies):
            try:
                data_fname = data_folder + meta_file[0:-4] + '_a.xim'
                file_data = np.loadtxt(data_fname) # for single energy
            except:
                data_fname = data_folder + meta_file[0:-4] + '_a{:03n}.xim'.format(i)
                file_data = np.loadtxt(data_fname) # for energy scans

            image_stack[i] = file_data
            
        return image_stack
    
    

    def __init__(self, data_folder, meta_file,
                 I0_meta_file= None , I0_file = None) :
        
        self.data_folder = data_folder
        self.meta_file = meta_file
        
        self.I0_meta_file = I0_meta_file
        self.I0_file = I0_file
        
        print('STXM class created at: \n' + str(self))
        
        print('Loading metadata ...')
        self.x_pos, self.y_pos, self.energies = self.get_stack_coordinates(self.data_folder +self.meta_file )
        print('done')
        
        print('loading image files ... ')
        self.image_stack = self.load_data(self.data_folder, self.meta_file,
                                          self.x_pos, self.y_pos, self.energies)
        print('done')


```

## Calculate XMCD

```python
# XMCD = -np.log(stxm_N_pol.image_stack_norm[idx_res]) - (-np.log(stxm_P_pol.image_stack_norm[idx_res]) )
XMCD = stxm_N_pol.image_stack_norm[idx_res] - stxm_P_pol.image_stack_norm[idx_res]
XMCD -= XMCD.mean()


fig, ax = plt.subplots(1, figsize = (10,8))
im = plt.pcolormesh(stxm_N_pol.x_pos, stxm_N_pol.y_pos, XMCD, cmap = 'seismic',
                   vmin = -0.2, vmax = 0.2)
cbar = plt.colorbar()
# ax.xticks()
ax.set_ylabel('Position ($\mu$m)', fontsize = 16)
ax.set_xlabel('Position ($\mu$m)', fontsize = 16)
cbar.set_label('XMCD', fontsize = 15)
```

```python
'''
Create a mask so that we can select the regoins where the grains are magnetised parallel or anti-parallel to the 
polarisation vector. 
'''
spin_up_mask = np.where(XMCD < 0.01) 
spin_down_mask = np.where(XMCD > -0.01) 



test_mask_img = np.zeros_like(XMCD)
test_mask_img[spin_up_mask] = 1
test_mask_img[spin_down_mask] = -1

fig, ax = plt.subplots(1, figsize = (10,8))

im = ax.pcolormesh(stxm_N_pol.x_pos, stxm_N_pol.y_pos, test_mask_img, cmap = 'gray')#, vmin = vmin, vmax = vmax)
plt.colorbar(im, ax = ax)
ax.set_ylabel('Position ($\mu$m)', fontsize = 16)
ax.set_xlabel('Position ($\mu$m)', fontsize = 16)


```

Here we stop the analysis for single scans... we can also run the following part for obtaining an XAS/XMCD curve. 

Now, the next part is to calculate an XAS spectrum, for that we need an energy stack

```python
if (len(stxm_N_pol.energies) > 1)&((stxm_N_pol.energies == stxm_P_pol.energies).all()):

    XMCD_stack = np.zeros_like(stxm_N_pol.image_stack_norm)
    for idx, energy in enumerate(stxm_N_pol.energies):
        XMCD_stack[idx] = (stxm_N_pol.image_stack_norm[idx] - stxm_P_pol.image_stack_norm[idx])
        XMCD_stack[idx] -= XMCD_stack[idx].mean()


    # There is a change in magnetic contrast, we are legit ... 
    fig, ax = plt.subplots(1, figsize = (10,8))

    im = plt.pcolormesh(stxm_N_pol.x_pos, stxm_N_pol.y_pos, XMCD_stack[20], cmap = 'seismic',
    # im = plt.pcolormesh(stxm_N_pol.x_pos, stxm_N_pol.y_pos, XMCD_stack[48], cmap = 'seismic',
    # im = plt.pcolormesh(stxm_N_pol.x_pos, stxm_N_pol.y_pos, XMCD_stack[62], cmap = 'seismic',

                        vmin = -0.06, vmax = 0.06
                       )
    cbar = plt.colorbar()
    # ax.xticks()
    ax.set_ylabel('Position ($\mu$m)', fontsize = 16)
    ax.set_xlabel('Position ($\mu$m)', fontsize = 16)
    cbar.set_label('XMCD', fontsize = 15)    

    
    stxm_N_pol.I_array_up = np.zeros(len(stxm_N_pol.energies))
    stxm_N_pol.I_array_down = np.zeros(len(stxm_N_pol.energies))
    stxm_P_pol.I_array_up = np.zeros(len(stxm_P_pol.energies))
    stxm_P_pol.I_array_down = np.zeros(len(stxm_P_pol.energies))

    # Normalised data 
    stxm_N_pol.I_array_up_norm = np.zeros(len(stxm_N_pol.energies))
    stxm_N_pol.I_array_down_norm = np.zeros(len(stxm_N_pol.energies))
    stxm_P_pol.I_array_up_norm = np.zeros(len(stxm_P_pol.energies))
    stxm_P_pol.I_array_down_norm = np.zeros(len(stxm_P_pol.energies))



    for i, energy in enumerate(stxm_N_pol.energies):

        stxm_N_pol.I_array_up[i] = stxm_N_pol.image_stack[i][spin_up_mask].mean()
        stxm_N_pol.I_array_down[i] = stxm_N_pol.image_stack[i][spin_down_mask].mean()
        stxm_P_pol.I_array_up[i] = stxm_P_pol.image_stack[i][spin_up_mask].mean()
        stxm_P_pol.I_array_down[i] = stxm_P_pol.image_stack[i][spin_down_mask].mean()
        
        # Normalised data
        stxm_N_pol.I_array_up[i] = stxm_N_pol.image_stack[i][spin_up_mask].mean()
        stxm_N_pol.I_array_down[i] = stxm_N_pol.image_stack[i][spin_down_mask].mean()
        stxm_P_pol.I_array_up[i] = stxm_P_pol.image_stack[i][spin_up_mask].mean()
        stxm_P_pol.I_array_down[i] = stxm_P_pol.image_stack[i][spin_down_mask].mean()
        
    '''
    We see here that there is a different absorption dependent on the magnetisation state in the sample. 
    '''

    plt.figure(figsize = (8,6))
    plt.plot(stxm_P_pol.energies, stxm_P_pol.I_array_down, label = 'Circular Postive Polarisation, spin down ROI')
    plt.plot(stxm_P_pol.energies, stxm_P_pol.I_array_up, label = 'Circular Postive Polarisation, spin up ROI')

    plt.plot(stxm_N_pol.energies, stxm_N_pol.I_array_up, label = 'Circular Negative Polarisation, spin up ROI' )
    plt.plot(stxm_N_pol.energies, stxm_N_pol.I_array_down, label = 'Circular Negative Polarisation, spin down ROI' )
    
    plt.legend(frameon = False)
    plt.title('Transmitted intensity (raw) in 2 different ROIs')
    '''
    Now, we can transform transmitted intensity into XAS, we will have to guess a background/I0 as 
    we measured I0 in a later measurement, this doesn't work (check stxm_N_pol.I_array_down_norm) 
    '''
    
    e_lim = np.max(np.where(stxm_P_pol.energies < 703))

    a_P1, b_P1 = np.polyfit(stxm_P_pol.energies[-3::], -np.log(stxm_P_pol.I_array_up)[-3::], 1)
    a_P2, b_P2 = np.polyfit(stxm_P_pol.energies[-3::], -np.log(stxm_P_pol.I_array_down)[-3::], 1)

    XAS_P1 = -np.log(stxm_P_pol.I_array_up) - (a_P1*stxm_P_pol.energies + b_P1)
    XAS_P2 = -np.log(stxm_P_pol.I_array_down) - (a_P2*stxm_P_pol.energies + b_P2)

    XAS_P1 -= XAS_P1[:e_lim].mean()
    XAS_P2 -= XAS_P2[:e_lim].mean()

    XAS_P1 /= XAS_P1[-3:].mean()
    XAS_P2 /= XAS_P2[-3:].mean() 
    
    
    fig, [ax1, ax2] = plt.subplots(2, sharex = True, figsize = (8,6))
    ax1.plot(stxm_N_pol.energies, XAS_P1, '.-', label = 'pseudo XAS in Red areas')
    ax1.plot(stxm_N_pol.energies, XAS_P2, '.-', label = 'pseudo XAS in Blue areas')
    ax1.legend(frameon = False)
    # ax1.set_xlabel('Energy (eV)')
    ax1.set_ylabel('XAS')
    ax1.set_title('Pseudo-XAS (no I0 normlisation)')
    # ax1.axvline(energies_P[e_lim])
    # ax1.axvline(energies_P[-4])

    ax2.plot(stxm_N_pol.energies, XAS_P1 -XAS_P2, '.-', label = 'pseudo XMCD')
    ax2.legend(frameon = False)
    plt.axhline(0, c='k', ls = '--')
    ax2.set_ylabel('XMCD')
    ax2.set_xlabel('energy (ev)')
    
else:
    print('there is only a single energy loaded... cannot do spectroscopy with a single energy...')
```
